```python
#Define Functions
def describe_employee(employee):
    """Prints detailed information about the employee."""
    print(f"Employee Name: {employee['name']}")
    print(f"Employee Number: {employee['number']}")
    print(f"Address: {employee['address']}")
    print(f"Salary: {employee['salary']}")
    print(f"Post: {employee['post']}")
    print(f"Year of Joining: {employee['year_of_joining']}")
    print(f"Education: {employee['education']}")
    print(f"Years of Experience: {employee['years_of_experience']}")
    print("-" * 40)

def update_year_of_experience(employee, additional_years):
    """Updates the years of experience of the employee."""
    if additional_years >= 0:
        employee['years_of_experience'] += additional_years
    else:
        print("Additional years cannot be negative.")

def increase_salary(employee, amount):
    """Increases the salary of the employee by the specified amount."""
    if amount > 0:
        employee['salary'] += amount
    else:
        print("Increment amount must be positive.")
```
